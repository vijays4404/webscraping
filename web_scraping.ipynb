```python
r=requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
```

```python
print(r)
```

```python
webpage=bs(r.content)
print(webpage)
```

```python
links=webpage.select('ul.socials a')
actual_links=[link['href'] for link in links]
print(actual_links)
```

```python
ulist=webpage.find('ul',attrs={'class':'socials'})
links=ulist.find_all('a')
actual_links=[link['href'] for link in links]
print(actual_links)
```

```python
table=webpage.select('table.hockey-stats')[0]
table
```

```python
column=table.find_all('thead')
column=table.find_all('th')
print(column)
```

```python
column_names=[c.string for c in column]
column_names
```

```python
table_rows=table.find('tbody').find_all('tr')
i=[]
for tr in table_rows:
    td=tr.find_all('td')
    row=[str(tr.string).strip() for tr in td]
    i.append(row)
i
```

```python
import pandas as pd
df=pd.DataFrame(i,columns=column_names)
print(df)
```

```python
df.loc[df['Team']!='Did not play']
```

```python
import re
facts=webpage.select('ul.fun-facts li')
facts=[fact.find(string=re.compile('is')) for fact in facts]
print(facts)
```

```python
facts_with_is=[fact.find_parent().get_text() for fact in facts if fact]
facts_with_is
```
